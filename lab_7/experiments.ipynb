# Model Aktywacji Progowej

## Zadanie 1

```python
from igraph import *
import random
import numpy as np
import matplotlib.pyplot as plt
from linear_threshold import simulation
```

```python
g = Graph.Watts_Strogatz(dim=1, size=20, nei=4, p=0.8)
g_history, activated_nodes_counts = simulation(g, m=5, n=0.3)
plt.plot(activated_nodes_counts)
plt.xticks(np.arange(len(activated_nodes_counts)))
plt.title("Liczba aktywowanych węzłów w iteracjach")
plt.xlabel("Iteracja")
plt.ylabel("Liczba aktywowanych węzłów")
plt.savefig("output/zad1_plot.eps")

for idx, g in enumerate(g_history):
    plot(g).save(f"./output/zad1_{idx}.png") 
```

## Zadanie 2

```python
num_nodes = 1000
n_values = [0.1, 0.3]
m_values = [int(num_nodes*0.05), int(num_nodes*0.1)]

fig, axes = plt.subplots(len(m_values), len(n_values), figsize=(10, 12))

idx = 0
for m in m_values:
    for n in n_values:
        g = Graph.Watts_Strogatz(dim=1, size=num_nodes, nei=2, p=0.8)
        g_history, activated_nodes_counts = simulation(g, m=m, n=n)
        ax = axes.flat[idx]
        ax.plot(activated_nodes_counts)
        ax.set_xticks(np.arange(len(activated_nodes_counts)))
        ax.set_title(f"m={m}, n={n}")
        ax.set_xlabel("Iteracja")
        ax.set_ylabel("Liczba aktywowanych węzłów")
        idx += 1
fig.tight_layout()
fig.savefig("output/zad2_plot.eps")
```
